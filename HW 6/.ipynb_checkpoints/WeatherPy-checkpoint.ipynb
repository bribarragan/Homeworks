```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

# Import API key
import api_keys

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)

cities[4]
```

## Perform API Calls

```python
# OpenWeatherMap API Key
api_key = api_keys.api_key

# Starting URL for Weather Map API Call
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + api_key + "&q="

cloudiness = []
country = []
date = []
humidity= []
lat= []
lng= []
max_temp= []
wind_speed= []
city_df=[]

for city in cities:
    request = requests.get(url+ city).json()
    try:
        cloudiness.append(request['clouds']['all'])
        country.append(request['sys']['country'])
        date.append(request['dt'])
        humidity.append(request['main']['humidity'])
        lat.append(request['coord']['lat'])
        lng.append(request['coord']['lon'])
        max_temp.append(request['main']['temp_max'])
        wind_speed.append(request['wind']['speed'])
        city_df.append(city)
        print(f'processing record {len(lat)} of {len(cities)} | {city}')
        print(url+city)
    except KeyError:
        print ('Missing City')
```

```python
print(f' cities = {len(cities)}, cloudiness = {len(cloudiness)}')

print(f' country = {len(country)}, date = {len(date)}')
print(f' humidity = {len(humidity)}, lat = {len(lat)}')
print(f' lng = {len(lng)}, max_temp = {len(max_temp)}, and WS is {len(wind_speed)}')
```

```python
df = pd.DataFrame({
    'City':city_df,
    'Cloudiness':cloudiness,
    'Country':country,
    'Date':date,
    'Humidity':humidity,
    'Lat': lat,
    'Lng': lng,
    'Max Temp': max_temp,
    'Wind Speed': wind_speed
    
})
```

```python
df.head()
df.to_csv('Cities Dataframe', sep='\t', encoding='utf-8')
```

```python
lat_temp = plt.plot(df['Lat'], df['Max Temp'], marker ='o', color='blue', linestyle='None')
plt.title('City Latitude vs Max Temperature (1/6/19)')
plt.xlabel('Latitude')
plt.ylabel('Max Temp')
plt.grid()
plt.xlim(-80,90)
plt.ylim(-100, 150)
plt.savefig('Lat vs Temp')
```

```python
lat_hum = plt.plot(df['Lat'], df['Humidity'], marker ='o', color='green', linestyle='None')
plt.title('City Latitude vs Humidity (1/6/19)')
plt.xlabel('Latitude')
plt.ylabel('Humidity %')
plt.grid()
plt.xlim(-80,90)
plt.ylim(-10, 110)
plt.savefig('Lat vs Humidity')
```

```python
lat_cloud = plt.plot(df['Lat'], df['Cloudiness'], marker ='o', color='red', linestyle='None')
plt.title('City Latitude vs Cloudiness (1/6/19)')
plt.xlabel('Latitude')
plt.ylabel('Cloudiness %')
plt.grid()
plt.xlim(-80,90)
plt.ylim(-10, 110)
plt.savefig('Lat vs Cloudiness')
```

```python
lat_wind = plt.plot(df['Lat'], df['Wind Speed'], marker ='o', color='green', linestyle='None')
plt.title('City Latitude vs Wind Speed (1/6/19)')
plt.xlabel('Latitude')
plt.ylabel('Wind Speed (mph)')
plt.grid()
plt.xlim(-80,90)
plt.ylim(0, 45)
plt.savefig('Lat vs Wind Speed')
```

3 observable trends. 
    1. If you move up the hemisphere towards a high latitude, you are likely have higher wind speeds.
    2. You also have lower max temperature in the higher region this time of year.
    3. Cloudiness doesn't seem to have a relationship to the latitude at this time.
    4. Humidity seems to currently be at the lowest in the southern region.
