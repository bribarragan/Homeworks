```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv(city_data_to_load)
ride_df= pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset

city_df.head()
ride_df.head()

merged_df = pd.merge(city_df, ride_df, on='city', how='outer')

# Display the data table for preview
merged_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types

urban = merged_df.loc[merged_df['type'] == 'Urban', :]
suburban = merged_df.loc[merged_df['type'] == 'Suburban', :]
rural = merged_df.loc[merged_df['type'] == 'Rural', :]


urban_x_axis = urban.groupby('city').agg({'type':'count'}) 
urban_y_axis = urban.groupby('city').agg({'fare': 'mean'})
u_drive = urban.drop_duplicates(subset='city', keep='first', inplace=False)
u_drive.set_index('city', inplace=True)
bb=u_drive['driver_count'].sort_index()


suburban_x_axis = suburban.groupby('city').agg({'type':'count'}) 
suburban_y_axis = suburban.groupby('city').agg({'fare': 'mean'})
s_drive = suburban.drop_duplicates(subset='city', keep='first', inplace=False)
s_drive.set_index('city', inplace=True)
suburban_driver_count=s_drive['driver_count'].sort_index()


rural_x_axis = rural.groupby('city').agg({'type':'count'}) 
rural_y_axis = rural.groupby('city').agg({'fare': 'mean'})
r_drive = rural.drop_duplicates(subset='city', keep='first', inplace=False)
r_drive.set_index('city', inplace=True)
rural_driver_count=r_drive['driver_count'].sort_index()


# Build the scatter plots for each city types


plt.grid()
plt.scatter(urban_x_axis, urban_y_axis, marker="o", facecolors="#33ccff", label="Urban", edgecolors="black",
            s=bb*10, alpha=0.5)
plt.scatter(suburban_x_axis, suburban_y_axis, marker="o", facecolors="#ffd700", label="Suburban", edgecolors="black",
            s=suburban_driver_count*10, alpha=0.5)
plt.scatter(rural_x_axis, rural_y_axis, marker="o", facecolors="#ffd8cc",  label="Rural", edgecolors="black",
            s=rural_driver_count*10, alpha=0.7)

# Incorporate the other graph properties

plt.title("Pyber Ride Sharing Data 2016")
plt.xlabel("Total Number or Rides per City")
plt.ylabel("Average Fare ($)")

# Create a legend
plt.legend()
# Incorporate a text label regarding circle size
plt.text(42,40,'Note: Circle size correlates with driver count')

# Save Figure
plt.savefig('Images/BubblePlot 2016 Data')
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
fares_by_type = merged_df.groupby('type')['fare'].sum()
pie = pd.DataFrame(fares_by_type)
pie.head()
explode = [.1, 0, 0]
colors = ['#ffd8cc', '#ffd700', '#33ccff']
# Build Pie Chart
plt.pie(pie['fare'], explode=explode, labels=pie.index, colors=colors, autopct="%1.1f%%", shadow=True, startangle=130)
plt.title('% of Total Fares by City Type')

# Save Figure
plt.savefig("Images/Fares_by_Type.png")
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
rides_by_type = merged_df.groupby('type')['type'].count()
ride = pd.DataFrame(rides_by_type)
pie.head()
explode = [0, .12, 0]
colors = ['#ffd8cc', '#ffd700', '#33ccff']
# Build Pie Chart
plt.pie(ride['type'], explode=explode, labels=ride.index, colors=colors, autopct="%1.1f%%", shadow=True, startangle=130)
plt.title('% of Total Rides by City Type')

# Save Figure
plt.savefig("Images/Rides_by_Type.png")
```

```python
# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents

drive_by_type = merged_df.drop_duplicates(subset='city', keep='first', inplace=False)
drivers = drive_by_type.groupby('type')['driver_count'].sum()
drivers = pd.DataFrame(drivers)
drivers.head()
explode = [.13, .12, 0]
colors = ['#ffd8cc', '#ffd700', '#33ccff']
# Build Pie Chart
plt.pie(drivers['driver_count'], explode=explode, labels=drivers.index, colors=colors, autopct="%1.1f%%", shadow=True, startangle=120)
plt.title('% of Total Drivers by City Type')

# Save Figure
plt.savefig("Images/Drivers_by_Type.png")
```

```python
# Show Figure
plt.show()
```
